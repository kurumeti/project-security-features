# ANALYSIST ON BANKNOTES SECURITY FEATURES

### Banknotes 
is a piece of paper money, constituting a central bank's promissory note to pay a stated sum to the bearer on demand. Banknotes were originally issued by commercial banks, which were legally required to redeem the notes for legal tender (usually gold or silver coin) when presented to the chief cashier of the originating bank. [source](https://en.wikipedia.org/wiki/Banknote) To prevent counterfeiting, it's important to secured Banknotes with the best security features.
<br>
<br>
Each country has their own concern on choosing the type of security features for its banknotes. Sometimes, the higher the denominations, the better security features on it. But, some country choose to standardize their banknotes security features, without concerning the denomination.
<br>
<br>

### The Most Favorite Security Features

```python
# change data type of Total from string to integer

df_max_sec['Total'] = df_max_sec['Total'].astype(int)
```

```python
# Find out the top 5 most used security features

sns.barplot(x = 'Total',
              y = 'Security Features',
              data = df_max_sec.nlargest(5, 'Total'))

# fig.9
```

**CONCLUSION**:
<br>

Based on the graph (fig.9), the most favorite types of security features used on banknotes, is **Blind Code** which is reasonable, because it also used to help the impaired vision to identify the notes. **Blind code** mostly print with **Intaglio**, a printing technique from a plate, where printing elements are incised into the surface. <br>

Followed by **Watermark**, which is a common security features, a faint design made in some paper during manufacture that is visible when held against the light and typically identifies the maker. <br>

**Rectoverso** and **Microtext** also 2 out 5 most used security features, which are also a common security features used in banknotes. <br>
